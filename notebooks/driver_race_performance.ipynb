```python
from fastf1.ergast import Ergast
import fastf1
import fastf1.plotting
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# The misc_mpl_mods option enables minor grid lines which clutter the plot
fastf1.plotting.setup_mpl(misc_mpl_mods=False)
```

```python
#Get Circuits Example
client = Ergast(result_type='pandas')
circuits = client.get_circuits(season=2023)
circuits
```

```python
#Get all Drivers:
drivers = client.get_driver_info(season=2023)
drivers
```

```python
#Get Lap Time Data:
all_lap_time_df = pd.DataFrame()
for eachDriver in drivers["driverId"]:
    total_results = False
    lap_times = client.get_lap_times(season=2023,round=16,driver=eachDriver,result_type='pandas')   
    while lap_times.is_complete == False:    
        if len(lap_times.content)>0:
            all_lap_time_df = all_lap_time_df.append(lap_times.content[0],ignore_index=True)
            try:
                lap_times=lap_times.get_next_result_page()
            except Exception as e:
                if str(e) == 'No more data after this response.':
                    break
```

```python
fig, axs = plt.subplots(figsize=(10,8))

sns.scatterplot(data=all_lap_time_df,x="number",y="time",ax=axs)
```

```python
## We will use race sessions to get the stint data for plotting:
race_round = 'Japan'
race = fastf1.get_session(2023, race_round,'R')
race.load()
```

```python
driver = '4'
driver_laps = race.laps.pick_driver(driver).pick_quicklaps().reset_index()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
sns.scatterplot(data=driver_laps,
                x="LapNumber",
                y="LapTime",
                ax=ax,
                hue="Compound",
                palette=fastf1.plotting.COMPOUND_COLORS,
                s=80,
                linewidth=0,
                legend='auto')

ax.set_xlabel("Lap Number")
ax.set_ylabel("Lap Time")

# The y-axis increases from bottom to top by default
# Since we are plotting time, it makes sense to invert the axis
ax.invert_yaxis()
# Turn on major grid lines
plt.grid(color='w', which='major', axis='both')
sns.despine(left=True, bottom=True)

plt.tight_layout()
plt.show()
```
